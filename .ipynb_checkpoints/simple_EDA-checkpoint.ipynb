```python
import pandas as pd
df = pd.read_csv('dataset/ml-latest-small/ratings.csv',sep=',')
```

```python
df
```

```python
df = df.assign(rating=1)
```

```python
df = df.drop(columns='ratings')
```

```python
df
```

```python
userdict={}
for index, row in df.iterrows():
    try:
        userdict[row['userId']] = userdict[row['userId']].append(row['movieId'])
    except KeyError:
        userdict.setdefault(row['userId'],[]).append(row['movieId'])
```
